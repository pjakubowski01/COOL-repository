```python
from scipy.optimize import linprog
import plotly.express as px

c = [10, 100]  # koszty jednostkowe dla A i B 

A_ub = [
    [2, 1],
    [1, 2] #lewy stron nierówności 
]
b_ub = [8, 10] #prawa stron nierówności

bounds = [(0.2, None), (0.2, None)] #ograniczenia jednostkowe dla argumentow x1 x2

res = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method="highs")

print(res.x)     # najlepsze wartości A i B
print(res.fun)   # minimalny koszt
print(res.success)
print(res.message)
```
